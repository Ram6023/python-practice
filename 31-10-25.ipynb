```python
column_names = ['Age','Education' ,'Martial_Status','Gender','Occupation','Income','Parent_Status','Country','Native','Weeks_Worked'] 
census_data = pd.read_csv("census.csv", names=column_names)

census_data.head()
```

```python
gender_count = census_data.groupby(['Gender']).size().reset_index(name='count')
gender_count
```

```python
edu_count = census_data.groupby(['Education']).size().reset_index(name='count')
edu_count
```
